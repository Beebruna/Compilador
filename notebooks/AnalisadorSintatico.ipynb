```python
class AnalisadorSintatico:
    def __init__(self, fonte):
        """
        Parâmetros
        ----------
        fonte: str
            nome do arquivo a ser analisado
        """
        self.al = AnalisadorLexico(fonte)
        self.pilha = [0]
        self.PARSER()
        
    def PARSER(self):
        token = self.al.SCANNER()
        a = token.classe  

        while True:
            s = self.pilha[-1]
            
            #print(f's={s}, a={a}')
            acao = tabela_sintatica[s][a]
            #print(f'{acao}, stack={self.pilha}')
            t = acao.numero        
                
            if acao.nome == 'shift':
                #print(f'Empilhando: {a} - {t}')
                self.pilha.append(t)
                token = self.al.SCANNER()
                a = token.classe 
        
            elif acao.nome == 'reduction':
                beta = betas[t]
                for i in range(beta):
                    self.pilha.pop()
                    
                A = producoes[t]['A'] # encontra A na produção A → Beta
                producao = producoes[t]['producao']
                t = self.pilha[-1]
                
                #print(f't={t}, A={A}')
                t=tabela_sintatica[t][A].numero
                self.pilha.append(t)
                print(f'Produção utilizada: {producao}')
            
            elif acao.nome == 'accept':
                break
                
            # Rotina de recuperação de erro
            else:
                pass
```
